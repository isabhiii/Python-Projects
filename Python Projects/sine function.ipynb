Plot the sine Function:

```python
# import the libraries
import matplotlib.pylab as plt
import numpy as np

# set the limit using numpy's linspace library.
x = np.linspace(-np.pi, np.pi, 201)     # 201 at the end defines the smoothness of curve, 15< smoothness< infinity

# plot the graph using np.sin() function
plt.plot(x, np.sin(x))
```

```python
# import the libraries
import matplotlib.pylab as plt
import numpy as np

# set the limit using numpy's linspace library.
x = np.linspace(-np.pi, np.pi, 201)     # 201 at the end defines the smoothness of curve, 15< smoothness< infinity

# plot the graph using np.sin() function
plt.plot(x, np.sin(x))

# assign the labels to the axis 
plt.xlabel('Angle [rad]')
plt.ylabel('sin(x)')
plt.axis('tight')
plt.show()
```
